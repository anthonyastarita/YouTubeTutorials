```python
def percentToDecimal(rawData):
    listData = [str(s) for s in rawData]
    
    if listData[-1] == '%':
        strData = "".join(listData[:-1])
    else:
        strData = "".join(listData)
    return float(strData)
```

```python
def meetsCriteria(ticker):
    
    try:
        stockInfo = finviz.get_stock(ticker)
        if percentToDecimal(stockInfo['Dividend %']) > 2.0 and percentToDecimal(stockInfo['EPS next 5Y']) > 8.0 and percentToDecimal(stockInfo['Sales past 5Y']) > 3.0:
            return True
    except:
        return False
    
    return False
```

```python
xls = pd.ExcelFile("tickerList.xls")
sheetX = xls.parse(0)
tickers = sheetX['Tickers']

for ticker in tickers:
    if meetsCriteria(ticker):
        print(ticker)
```
